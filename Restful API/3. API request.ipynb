```python
import requests
import pandas as pd
import json
```

```python
# GET ALL testing
base_url=r"http://127.0.0.1:5000"
response = requests.get(f"{base_url}/api/products")
print(response)

json_data = response.json()
df = pd.DataFrame(json_data)
print(response.content)
df
```

```python
# Function to fetch data for a specific Product ID
def fetch_product_by_id(product_id):
    # Construct the URL for fetching data for the specified Product ID
    url = f"{base_url}/api/products/{product_id}"
    
    # Send a GET request to the API endpoint
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Extract the JSON data from the response
        product_data = response.json()
        return product_data
    else:
        # Print error message if request was not successful
        print(f"Failed to fetch data for Product ID {product_id}. Status code: {response.status_code}")
        return None
```

```python
# POST testing 

# Define the payload
payload = {
  "User ID": 594,
  "Product ID": 1001,
  "Product Name": "Shoes",
  "Brand": "Adidas",
  "Category": "Women's Fashion",
  "Price": 693,
  "Rating": 2.0563252988,
  "Color": "Red",
  "Size": "S"
}

# Convert the payload to JSON format and set header
json_payload = json.dumps(payload)
headers = {'Content-Type': 'application/json'}

# Send the POST request with the JSON payload and headers
response = requests.post(f"{base_url}/api/products", data=json_payload, headers=headers)
print(response.content)
```

```python
# Example usage: Fetch data for Product ID 1001, new row added
product_id = '1001'
product_data = fetch_product_by_id(product_id)
if product_data:
    print(f"Data for Product ID {product_id}:")
    print(product_data)
```

```python
# DELETE TESTING

# Function to delete a specific product by its ID using the requests module
def delete_product(product_id):
    # Define the URL of your Flask API endpoint
    url = f"{base_url}/api/products/{product_id}"
    
    # Send a DELETE request to the API endpoint
    response = requests.delete(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Print success message if deletion was successful
        print(f"Product with ID {product_id} deleted successfully")
    else:
        # Print error message if deletion was not successful
        print(f"Failed to delete product with ID {product_id}. Status code: {response.status_code}")

# Example usage: Delete product with ID '1001'
product_id = '1001'
delete_product(product_id)
```
