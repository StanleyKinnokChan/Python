```python
#https://preppindata.blogspot.com/2021/02/2021-week-5-dealing-with-duplication.html

import pandas as pd
```

### Input the data

```python
df = pd.read_csv(r'data/PD 2021 Wk 5 Input.csv')
df['From Date'] = pd.to_datetime(df['From Date'],format='%d/%m/%Y')
df
```

### For each Client, work out who the most recent Account Manager is

```python
df_most_recent = df.sort_values('From Date').groupby(['Client ID']).tail(1)
df_most_recent = df_most_recent[['Client ID','Account Manager']]
df_most_recent.rename(columns= {'Account Manager': 'Recent Account Manager'}, inplace=True)
```

### Filter the data so that only the most recent Account Manager remains
- Be careful not to lose any attendees from the training sessions!
- In some instances, the Client ID has changed along with the Account Manager. Ensure only the most recent Client ID remains

```python
# join the main df with the most recent Account Manager for each client
df = df.merge(df_most_recent, on='Client ID', how = 'inner')
df
```

```python
#filter the data to only Recent Account Manager for each client
df = df[df['Account Manager'] == df['Recent Account Manager']]
df
```

### Output the data

```python
df.to_csv(r'output/2021-week5-output.csv')
```
