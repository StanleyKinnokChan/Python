```python
# https://preppindata.blogspot.com/2021/03/2021-week-12-maldives-tourism.html

import pandas as pd
import numpy as np
from datetime import datetime
```

### Input the data

```python
df_Tourism = pd.read_csv(r'C:\Users\StanleyChan\SynologyDrive\Work Backup\my document\python\Preppin Data\data\PD 2021 Wk 12 Input.csv')
df_Tourism.head()
```

### Pivot all of the month fields into a single column
-  Rename the fields and ensure that each field has the correct data type

```python
df_Tourism = df_Tourism.melt(id_vars=['id', 'Series-Measure', 'Hierarchy-Breakdown', 'Unit-Detail'],
                var_name='Month',
                value_name='Number of Tourists')

df_Tourism['Month'] = pd.to_datetime(df_Tourism['Month'], format='%b-%y').dt.strftime('%d/%m/%Y')
df_Tourism
```

```python
# remove % sign, replace 'na' as null, change data type
df_Tourism['Number of Tourists'] = df_Tourism['Number of Tourists'].str.replace('%', '')
df_Tourism['Number of Tourists'] = df_Tourism['Number of Tourists'].replace('na', np.nan).astype(float)

df_Tourism
```

### Filter out the nulls & Filter our dataset so our Values are referring to Number of Tourists

```python
df_Tourism = df_Tourism[~df_Tourism['Number of Tourists'].isnull()]
df_Tourism
df_Tourism = df_Tourism[df_Tourism['Unit-Detail'].str.contains('Tourists')]
# df_Tourism['Series-Measure'] = df_Tourism['Series-Measure'].str.extract('Tourist arrivals from (.*)',expand=True)
df_Tourism
```

### Our goal now is to remove all totals and subtotals from our dataset so that only the lowest level of granularity remains. 
- Currently we have Total > Continents > Countries, but we don't have data for all countries in a continent, so it's not as simple as just filtering out the totals and subtotals
- Plus in our Continents level of detail, we also have The Middle East and UN passport holders as categories.

```python
# first, see what measures we have
df_Tourism['Series-Measure'].unique()
```

```python
# filter out Total 
df_Tourism = df_Tourism[df_Tourism['Series-Measure'] != 'Total tourist arrivals']

# strip the prefix sentences
df_Tourism.loc[:,['Series-Measure']] = df_Tourism['Series-Measure'].str.replace('Tourist arrivals from ','')
df_Tourism.loc[:,['Series-Measure']] = df_Tourism['Series-Measure'].str.replace('Tourist arrivals - ','')


df_Tourism['Breakdown'] = df_Tourism['Hierarchy-Breakdown'].str.extract('Real Sector / Tourism / Tourist arrivals / (\w+)',expand=True)
df_Tourism

df_continent = df_Tourism[df_Tourism['Breakdown'].isnull()]
df_country = df_Tourism[~df_Tourism['Breakdown'].isnull()]

df_country = df_country.rename(columns={'Series-Measure':'Country'})
df_country = df_country[['Month','Breakdown','Country','Number of Tourists']]
df_country
```

```python
# agg the country to continent level
df_country_agg = df_country.groupby(['Breakdown','Month'])['Number of Tourists'].sum().reset_index()
df_country_agg
```

```python
# calculate remain tourists in the unknown countries
df_unknown = df_continent.merge(df_country_agg, left_on=['Series-Measure','Month'], right_on=['Breakdown','Month'], how='left')
df_unknown['Number of Tourists_y'] = np.where(df_unknown['Number of Tourists_y'].isna(),0,df_unknown['Number of Tourists_y'])
df_unknown['Number of Tourists'] = df_unknown['Number of Tourists_x'] - df_unknown['Number of Tourists_y']
df_unknown

#rename and select target column
df_unknown['Country'] = 'Unknown'
df_unknown = df_unknown.rename(columns={'Series-Measure':'Breakdown'})
df_unknown = df_unknown[['Month','Breakdown','Country','Number of Tourists']]
df_unknown
```

```python
df_Tourists_all = pd.concat([df_country,df_unknown])
df_Tourists_all
```

### Output the data

```python
df_Tourists_all.to_csv(r'output/2021-week12-output.csv')
```
