### Input Data


```python
df_worldcup = pd.read_excel(r'data\PD 2021 Wk 28 Input.xlsx', sheet_name='WorldCup')
df_euros = pd.read_excel(r'data\PD 2021 Wk 28 Input.xlsx', sheet_name='Euros')
df_worldcup
```

```python
#combine both df
df_main = pd.concat([df_worldcup.assign(source='WorldCup'), df_euros.assign(source='Euros')])
df_main
```

### Determine what competition each penalty was taken in


```python
df_main['Event'] = df_main['source'] + ' ' + df_main['Event Year'].astype(str) + ' ' + df_main['Round']
df_main
```

### Clean any fields, correctly format the date the penalty was taken, & group the two German countries (eg, West Germany & Germany)


```python
# remove , from Year
df_main['Event Year'] = df_main['Event Year'].str.replace(',', '')
df_main['Year'] = df_main['Event Year'] 
df_main.dtypes
```

```python
# correct the year of the date
df_main['Event Year'] = df_main['Event Year'].astype(int)
df_main['Month'] = df_main['Date'].dt.month
df_main['Day'] = df_main['Date'].dt.day
df_main['Date'] = pd.to_datetime(df_main[['Year', 'Month', 'Day']])
df_main.head(2)
```

```python
# clean the suffix
df_main['Winner'] = df_main['Winner'].str.strip()
df_main['Loser'] = df_main['Loser'].str.strip()
list(df_main['Loser'].unique())
```

```python
# clean the 'West ' so that 'West German' is 'German'
df_main['Winner'] = df_main['Winner'].str.replace('West ', '')
df_main['Loser'] = df_main['Loser'].str.replace('West ', '')
```

```python
df_main.columns
```

### Rank the countries on the following: 
- Shootout win % (exclude teams who have never won a shootout)
- Penalties scored %

```python
###Shootout win % (exclude teams who have never won a shootout) ######################################################################## 

df_S_win = df_main[['Date','Winner','Loser']]
df_S_win = df_S_win.drop_duplicates()
df_S_win = df_S_win.melt(
    id_vars=['Date'],
             var_name='Winner/Loser',
             value_name='Team' 
)
df_S_win = df_S_win.groupby(['Winner/Loser','Team'])['Date'].count().reset_index()
df_S_win = df_S_win.pivot(index='Team',columns='Winner/Loser',values='Date').reset_index()

# replace na as 0
df_S_win['Loser'] = np.where(df_S_win['Loser'].isna(),0,df_S_win['Loser'])
df_S_win['Winner'] = np.where(df_S_win['Winner'].isna(),0,df_S_win['Winner'])

# remove team that never won
df_S_win = df_S_win[df_S_win['Winner']>0]

# calculate the %
df_S_win['Total Shootouts'] = df_S_win['Winner']+df_S_win['Loser']
df_S_win['Shootout Win %'] = df_S_win['Winner']/df_S_win['Total Shootouts']*100

# rank and sort
df_S_win['Winner % Rank'] = df_S_win['Shootout Win %'].rank(ascending=False,method='dense').astype(int)
df_S_win = df_S_win.sort_values(by='Winner % Rank')

# Convert columns to int & rename winner to shootouts
df_S_win[['Winner', 'Total Shootouts', 'Shootout Win %']] = df_S_win[['Winner', 'Total Shootouts', 'Shootout Win %']].astype(int)
df_S_win.rename(columns={'Winner':'Shootouts'}, inplace=True)

# reorder columns
df_S_win = df_S_win[['Winner % Rank','Shootout Win %','Total Shootouts','Shootouts','Team']]

# ready for output
df_S_win
```

```python
df_main['Winner scored?'] = df_main['Winning team Taker'].str.contains('scored')
df_main['Loser scored?'] = df_main['Losing team Taker'].str.contains('scored')

df_main['Winning team penalty score'] = np.where(df_main['Winner scored?'] == True, 1, np.where(df_main['Winner scored?'] == False,0,None))
df_main['Losing team penalty score'] = np.where(df_main['Loser scored?'] == True, 1, np.where(df_main['Loser scored?'] == False,0,None))

df_pen_winer = df_main[['Event','Winner','Penalty Number ','Winning team penalty score']]
df_pen_lower = df_main[['Event','Loser','Penalty Number ','Losing team penalty score']]

columns_name = ['Event','Team','penalty number','Score']
df_pen_winer.columns = columns_name
df_pen_lower.columns = columns_name

df_pen_winer[df_pen_winer['Team']=='Russia']
df_pen_lower[df_pen_lower['Team']=='Russia']

df_pen_total = pd.concat([df_pen_winer,df_pen_lower])
df_pen_total = df_pen_total[~df_pen_total['Score'].isna()]
df_pen_total[df_pen_total['Team']=='Russia']
```

```python

df_pen_pct = df_pen_total.groupby(['Team'],as_index=False).agg({'Score':['sum','size']})
df_pen_pct.columns = list(map(''.join, df_pen_pct.columns.values))

df_pen_pct['Penalties Missed'] = df_pen_pct['Scoresize'] - df_pen_pct['Scoresum']
df_pen_pct['% Total Penalties Scored'] = df_pen_pct['Scoresum']/ df_pen_pct['Scoresize'] * 100
df_pen_pct

# rename column, set int datatype
df_pen_pct.columns=['Team','Penalties Scored','Total','Penalties Missed','% Total Penalties Scored']
df_pen_pct['% Total Penalties Scored'] = df_pen_pct['% Total Penalties Scored'].astype(float).round(0).astype(int)
df_pen_pct = df_pen_pct.sort_values(by='% Total Penalties Scored', ascending=False)
df_pen_pct

# rank and reorder
df_pen_pct['Penalties Scored %Rank'] = df_pen_pct['% Total Penalties Scored'].rank(ascending=False,method='dense').astype(int)
df_pen_pct = df_pen_pct[['Penalties Scored %Rank','% Total Penalties Scored', 'Penalties Missed', 'Penalties Scored','Team']]

df_pen_pct
```

### What is the most and least successful time to take a penalty? (What penalty number are you most likely to score or miss?)


```python
df_pen_total.head()
```

```python
df_pen_num= df_pen_total.groupby(['Team'])['Score'].agg(['sum','count']).reset_index()

df_pen_num['% Total Penalties Scored'] = df_pen_num['sum']/ df_pen_num['count'] * 100
df_pen_num['% Total Penalties Scored'] = df_pen_num['% Total Penalties Scored'].astype(float).round(0).astype(int)
df_pen_num['Penalties Missed'] = df_pen_num['count'] - df_pen_num['sum']
df_pen_num


df_pen_num['Penalties Scored %Rank'] = df_pen_num['% Total Penalties Scored'].rank(method='dense',ascending=False).astype(int)

df_pen_num = df_pen_num.sort_values(by='Penalties Scored %Rank', ascending=True).reset_index()
df_pen_num.rename(columns={'sum':'Penalties Scored'},inplace=True)
df_pen_num = df_pen_num[['Penalties Scored %Rank','% Total Penalties Scored','Penalties Missed','Penalties Scored','Team']]
df_pen_num
```
