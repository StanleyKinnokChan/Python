```python
#https://preppindata.blogspot.com/2021/01/2021-week-1.html

import pandas as pd
import numpy as np
```

### Connect and load the csv file

```python
df = pd.read_csv(r'data\PD 2021 Wk 1 Input - Bike Sales.csv')
```

```python
df
```

### Split the 'Store-Bike' field into 'Store' and 'Bike

```python
df[['Store','Bike']] = df['Store - Bike'].str.split(' - ', expand=True)
```

### clean up the 'Bike' field to leave just three values in the 'Bike' field (Mountain, Gravel, Road)



```python
df['Bike'].unique()
```

```python
df['Bike']=df['Bike'].str[0]
df['Bike'] = np.where(df['Bike'] == 'M','Mountain', np.where(df['Bike']== 'R', 'Road', 'Gravel'))
```

```python
df['Bike'].unique()
```

### Create two different cuts of the date field: 'quarter' and 'day of month' 

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Quarter'] = df['Date'].dt.quarter 
df['Day of Month'] = df['Date'].dt.day
df
```

### Remove the first 10 orders

```python
df = df[df['Order ID'] >= 11]
df
```

### Output the data as a csv

```python
df = df.drop(['Store - Bike','Date'], axis=1)
df.to_csv(r'output/2021-week1-output.csv')
```
